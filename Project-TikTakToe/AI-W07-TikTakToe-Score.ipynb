```python
def  printState(xState):
  print("\n\n\n _____________________________ ")
  #print(xState)
  #print("\n ....... ")
  for row in xState:
    s = ""
    for cell in row:
      curr = "-"
      if (cell == 1):
        curr = "X"
      if (cell == 2):
        curr = "O"
      s += "\t" + curr
    print(s)
  print("\n ....... ")

def  inputOPosition():
  r = int ( input("Nhap ROW (<N): ") )
  c = int ( input("Nhap COLUMN (<N): ") )
  return (r,c)

def  inputXY():
  inXY = input("Nhap vi tri can Danh: ROW COLUMN (<N): ")
  inXY = inXY.strip().split()
  r = int ( inXY[0] )
  c = int ( inXY[1] )
  return (r,c)
```

```python
def  isWinner(x):
  global N
  global currentState
  nWin = True 
  ## ROWs
  i = 0
  while ( nWin and (i < N) ):
    sl = 0
    for j in range(N):
      if (currentState[i,j] == x):
        sl += 1
    if (sl == N):
      nWin = False 
    i += 1
  
  ## Columns
  i = 0
  while ( nWin and (i < N) ):
    sl = 0
    for j in range(N):
      if (currentState[j,i] == x):
        sl += 1
    if (sl == N):
      nWin = False 
    i += 1
  
  ## cheo 1
  sl = 0
  for j in range(N):
    if (currentState[j,j] == x):
      sl += 1
  if (sl == N):
    nWin = False 
  
  ## cheo 2
  sl = 0
  for j in range(N):
    if (currentState[j,N-1-j] == x):
      sl += 1
  if (sl == N):
    nWin = False 

  return not nWin
  
```

```python
## check 

def checkRCvalid(r,c):
  flag = False
  if (0 <= r  and r < N and 0<=c and c<N):
    flag = True
  return flag

def checkAvailable(r,c):
  flag = False
  if (currentState[r,c] == 0):
    flag = True
  return flag
```

## Tính SCORE (Heuristic - H function) cho giải pháp 

```python
# Score H01 - tính dựa trên 
# - Số cơ hội 
# + số nước đi ngắn nhất 

## Hiện thực cho Matrix 3x3
def  checkOneLine(valueArr,x):
    scoreOne = (0, 1000)
    if ( valueArr == (x,0,0) ):
        scoreOne = (1, 2)
    elif ( valueArr == (x,x,0) ):
        scoreOne = (1, 1) 
    elif ( valueArr == (x,x,x) ):
        scoreOne = (1, 0)
    return scoreOne
    
def  calculateSCORE_H01(solution, x):
    #--- Row 0
    score01 = checkOneLine( 
        ( solution[0,0], solution[0,1], solution[0,2] ) , x)
    minSteps = score01[1]
    totalSols = score01[0]
    #--- Row 1
    score01 = checkOneLine( 
        ( solution[1,0], solution[1,1], solution[1,2] ) , x)
    if (minSteps > score01[1]):
        minSteps = score01[1]
    totalSols += score01[0]
    #--- Row 2
    score01 = checkOneLine( 
        ( solution[2,0], solution[2,1], solution[2,2] ) , x)
    if (minSteps > score01[1]):
        minSteps = score01[1]
    totalSols += score01[0]
    #--- Col 0
    score01 = checkOneLine( 
        ( solution[0,0], solution[1,0], solution[2,0] ) , x)
    if (minSteps > score01[1]):
        minSteps = score01[1]
    totalSols += score01[0]
    #--- Col 1
    score01 = checkOneLine( 
        ( solution[0,1], solution[1,1], solution[2,1] ) , x)
    if (minSteps > score01[1]):
        minSteps = score01[1]
    totalSols += score01[0]
    #--- Col 2
    score01 = checkOneLine( 
        ( solution[0,2], solution[1,2], solution[2,2] ) , x)
    if (minSteps > score01[1]):
        minSteps = score01[1]
    totalSols += score01[0]
    #--- diagonal 1
    score01 = checkOneLine( 
        ( solution[0,0], solution[1,1], solution[2,2] ) , x)
    if (minSteps > score01[1]):
        minSteps = score01[1]
    totalSols += score01[0]
    #--- diagonal 2
    score01 = checkOneLine( 
        ( solution[2,0], solution[1,1], solution[0,2] ) , x)
    if (minSteps > score01[1]):
        minSteps = score01[1]
    totalSols += score01[0]
    return  (totalSols, minSteps)
```

```python
def  selectSolution():
    #r = random.randint(0, N)
    #c = random.randint(0, N)
    bestSols = None
    for r in range(N):
        for c in range(N):
            if (checkRCvalid(r,c)):
                if ( checkAvailable(r,c) ):
                    ##### currentState[r,c] = 1
                    newState = currentState.copy()
                    newState[r,c] = 1
                    scoreSol = calculateSCORE_H01(newState, 2)
                    if (bestSols == None):
                        bestSols = { "score" : scoreSol,
                                    "pos" : (r,c) }
                    else:
                        if (bestSols["score"][1] < scoreSol[1]):
                            bestSols = { "score" : scoreSol,
                                    "pos" : (r,c) }
    if (bestSols != None):
        r = bestSols["pos"][0]
        c = bestSols["pos"][1]
        currentState[r,c] = 1
        print("... Chon nuoc di ", bestSols)
    else:
        print(".... khong chon duoc giai phap !!!")
```

## Main Process

```python
## Zero State 
currentState =   np.zeros(( N, N ))  # tiktaktoeBoard


## Playing
printState(currentState)

while ( not (isWinner(1) or isWinner(2)) ):   
  ## Máy chơi
  selectSolution()  # Máy chọn nước đi kế tiếp
  printState(currentState)

  ## Người chơi
  r,c = inputXY()   # inputOPosition()
  if (checkRCvalid(r,c)):
    if ( checkAvailable(r,c) ):
      currentState[r,c] = 2
      printState(currentState)
    else:
      print(" ....... vi tri da DANH ! ")
  else:
    print(" ...... r,c NGOAI pham vi ")


if isWinner(1):
    print("Player 1 la Nguoi THANG")
if isWinner(2):
    print("Player 2 la Nguoi THANG")
```
